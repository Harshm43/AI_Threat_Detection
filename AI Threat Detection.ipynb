```python
## Week of Apr 1–7: Baseline Autoencoder Model

import pandas as pd

# Column names from official NSL-KDD documentation
columns = [
    'duration','protocol_type','service','flag','src_bytes','dst_bytes','land','wrong_fragment','urgent',
    'hot','num_failed_logins','logged_in','num_compromised','root_shell','su_attempted','num_root',
    'num_file_creations','num_shells','num_access_files','num_outbound_cmds','is_host_login','is_guest_login',
    'count','srv_count','serror_rate','srv_serror_rate','rerror_rate','srv_rerror_rate','same_srv_rate',
    'diff_srv_rate','srv_diff_host_rate','dst_host_count','dst_host_srv_count','dst_host_same_srv_rate',
    'dst_host_diff_srv_rate','dst_host_same_src_port_rate','dst_host_srv_diff_host_rate','dst_host_serror_rate',
    'dst_host_srv_serror_rate','dst_host_rerror_rate','dst_host_srv_rerror_rate','class','difficulty'
]

# Load training and test sets
train_url = "https://raw.githubusercontent.com/defcom17/NSL_KDD/master/KDDTrain+.txt"
test_url = "https://raw.githubusercontent.com/defcom17/NSL_KDD/master/KDDTest+.txt"

train_df = pd.read_csv(train_url, names=columns)
test_df = pd.read_csv(test_url, names=columns)

# Drop 'difficulty' column (not needed for modeling)
train_df.drop('difficulty', axis=1, inplace=True)
test_df.drop('difficulty', axis=1, inplace=True)

# Show basic info
print("Training set shape:", train_df.shape)
print("Testing set shape:", test_df.shape)
train_df.head()

# Binary labels: 0 = normal, 1 = attack
train_df['label'] = train_df['class'].apply(lambda x: 0 if x == 'normal' else 1)
test_df['label'] = test_df['class'].apply(lambda x: 0 if x == 'normal' else 1)

from sklearn.preprocessing import LabelEncoder

# Encode categorical columns
categorical_cols = ['protocol_type', 'service', 'flag']
encoder = LabelEncoder()

for col in categorical_cols:
    train_df[col] = encoder.fit_transform(train_df[col])
    test_df[col] = encoder.transform(test_df[col])

from sklearn.preprocessing import MinMaxScaler

# Drop the original class column
train_df.drop('class', axis=1, inplace=True)
test_df.drop('class', axis=1, inplace=True)

# Separate features and label
X_train = train_df.drop('label', axis=1)
y_train = train_df['label']
X_test = test_df.drop('label', axis=1)
y_test = test_df['label']

# Normalize
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

from tensorflow import keras

Model = keras.Model
Input = keras.layers.Input
Dense = keras.layers.Dense
EarlyStopping = keras.callbacks.EarlyStopping

# Build autoencoder
input_dim = X_train_scaled.shape[1]
input_layer = Input(shape=(input_dim,))
encoded = Dense(32, activation="relu")(input_layer)
encoded = Dense(16, activation="relu")(encoded)
encoded = Dense(8, activation="relu")(encoded)

decoded = Dense(16, activation="relu")(encoded)
decoded = Dense(32, activation="relu")(decoded)
output_layer = Dense(input_dim, activation="sigmoid")(decoded)

autoencoder = Model(inputs=input_layer, outputs=output_layer)
autoencoder.compile(optimizer="adam", loss="mse")

# Train only on normal data (label 0)
X_train_normal = X_train_scaled[y_train == 0]

# Fit the model
history = autoencoder.fit(
    X_train_normal, X_train_normal,
    epochs=20,
    batch_size=256,
    shuffle=True,
    validation_split=0.1,
    callbacks=[EarlyStopping(patience=3)],
    verbose=1
)

import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

# Get reconstructions
reconstructions = autoencoder.predict(X_test_scaled)
mse = np.mean(np.power(X_test_scaled - reconstructions, 2), axis=1)

# Choose a threshold (e.g. 95th percentile of normal training errors)
threshold = np.percentile(mse[y_test == 0], 95)
print("Reconstruction error threshold:", threshold)

# Predict: 1 = anomaly, 0 = normal
y_pred = [1 if e > threshold else 0 for e in mse]

# Evaluate
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(mse[y_test == 0], bins=50, color='green', label='Normal', kde=True)
sns.histplot(mse[y_test == 1], bins=50, color='red', label='Attack', kde=True)
plt.axvline(threshold, color='blue', linestyle='--', label='Threshold')
plt.title('Reconstruction Error by Class')
plt.xlabel('MSE')
plt.ylabel('Frequency')
plt.legend()
plt.show()

from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Normal','Attack'], yticklabels=['Normal','Attack'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

from sklearn.metrics import roc_curve, auc

fpr, tpr, _ = roc_curve(y_test, mse)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()


##Apr 8–14: Train anomaly detection model (Autoencoder or LSTM)

import numpy as np

# LSTM expects 3D input: (samples, time_steps, features)
X_train_lstm = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_lstm = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, RepeatVector, TimeDistributed, Dense

model_lstm = Sequential([
    LSTM(64, activation='relu', input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2]), return_sequences=True),
    LSTM(32, activation='relu', return_sequences=False),
    RepeatVector(1),
    LSTM(32, activation='relu', return_sequences=True),
    LSTM(64, activation='relu', return_sequences=True),
    TimeDistributed(Dense(X_train_lstm.shape[2]))
])

model_lstm.compile(optimizer='adam', loss='mse')
model_lstm.summary()

# Train only on normal data
X_train_lstm_normal = X_train_lstm[y_train == 0]

history = model_lstm.fit(
    X_train_lstm_normal, X_train_lstm_normal,
    epochs=20,
    batch_size=256,
    validation_split=0.1,
    shuffle=True,
    verbose=1
)

from sklearn.metrics import classification_report, confusion_matrix

# Get reconstructions on test set
reconstructions_lstm = model_lstm.predict(X_test_lstm)
mse_lstm = np.mean(np.power(X_test_lstm - reconstructions_lstm, 2), axis=(1,2))

# Pick a threshold (same strategy: 95th percentile on normal samples)
threshold_lstm = np.percentile(mse_lstm[y_test == 0], 95)
print("Reconstruction error threshold (LSTM):", threshold_lstm)

# Make predictions
y_pred_lstm = [1 if e > threshold_lstm else 0 for e in mse_lstm]

# Report
print(classification_report(y_test, y_pred_lstm))
print(confusion_matrix(y_test, y_pred_lstm))

plt.figure(figsize=(10,5))
sns.histplot(mse_lstm[y_test == 0], bins=50, color='green', label='Normal', kde=True)
sns.histplot(mse_lstm[y_test == 1], bins=50, color='red', label='Attack', kde=True)
plt.axvline(threshold_lstm, color='blue', linestyle='--', label='Threshold')
plt.title('LSTM Reconstruction Error by Class')
plt.xlabel('MSE')
plt.ylabel('Frequency')
plt.legend()
plt.show()

cm_lstm = confusion_matrix(y_test, y_pred_lstm)
plt.figure(figsize=(5,4))
sns.heatmap(cm_lstm, annot=True, fmt="d", cmap="Purples", xticklabels=['Normal','Attack'], yticklabels=['Normal','Attack'])
plt.title("LSTM Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

fpr_lstm, tpr_lstm, _ = roc_curve(y_test, mse_lstm)
roc_auc_lstm = auc(fpr_lstm, tpr_lstm)

plt.figure(figsize=(6, 5))
plt.plot(fpr_lstm, tpr_lstm, color='orange', lw=2, label=f'LSTM ROC (AUC = {roc_auc_lstm:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('LSTM Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

## 🧬 Apr 15–21: Train GAN to Generate Synthetic Attacks

import pandas as pd
import numpy as np

# Use preprocessed and normalized data
X_train_attacks = X_train_scaled[y_train == 1]  # Only attack samples
print("Attack samples:", X_train_attacks.shape[0])

import tensorflow as tf
from tensorflow.keras.layers import Dense, LeakyReLU, BatchNormalization
from tensorflow.keras.models import Sequential

# Define generator
def build_generator(latent_dim, output_dim):
    model = Sequential()
    model.add(Dense(128, input_dim=latent_dim))
    model.add(LeakyReLU(alpha=0.2))
    model.add(BatchNormalization())
    model.add(Dense(256))
    model.add(LeakyReLU(alpha=0.2))
    model.add(BatchNormalization())
    model.add(Dense(output_dim, activation='tanh'))  # Range -1 to 1
    return model

# Define discriminator
def build_discriminator(input_dim):
    model = Sequential()
    model.add(Dense(256, input_dim=input_dim))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dense(128))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dense(1, activation='sigmoid'))
    return model

# Normalize attack data to [-1, 1] (required for 'tanh')
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1, 1))
X_attack_scaled = scaler.fit_transform(X_train_attacks)

latent_dim = 32
input_dim = X_attack_scaled.shape[1]

# Build models
generator = build_generator(latent_dim, input_dim)
discriminator = build_discriminator(input_dim)
discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# GAN combines generator + discriminator
discriminator.trainable = False
gan_input = tf.keras.Input(shape=(latent_dim,))
fake_sample = generator(gan_input)
gan_output = discriminator(fake_sample)
gan = tf.keras.Model(gan_input, gan_output)
gan.compile(optimizer='adam', loss='binary_crossentropy')

epochs = 200
batch_size = 64
half_batch = batch_size // 2

for epoch in range(epochs):

    # Train discriminator
    idx = np.random.randint(0, X_attack_scaled.shape[0], half_batch)
    real_samples = X_attack_scaled[idx]

    noise = np.random.normal(0, 1, (half_batch, latent_dim))
    generated_samples = generator.predict(noise)

    d_loss_real = discriminator.train_on_batch(real_samples, np.ones((half_batch, 1)))
    d_loss_fake = discriminator.train_on_batch(generated_samples, np.zeros((half_batch, 1)))
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    # Train generator
    noise = np.random.normal(0, 1, (batch_size, latent_dim))
    valid_y = np.ones((batch_size, 1))  # wants to fool discriminator
    g_loss = gan.train_on_batch(noise, valid_y)

    if epoch % 500 == 0:
        print(f"{epoch} [D loss: {d_loss[0]:.4f}, acc: {d_loss[1]*100:.2f}%] [G loss: {g_loss:.4f}]")
# Generate new samples
num_samples = 500
noise = np.random.normal(0, 1, (num_samples, latent_dim))
generated_attacks = generator.predict(noise)

# Inverse transform to original scale
synthetic_attacks = scaler.inverse_transform(generated_attacks)

print("Synthetic Attack Sample (first 5):")
print(synthetic_attacks[:5])

##Apr 22–28: Evaluate both models and compare results

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Autoencoder
print("Autoencoder Classification Report")
print(classification_report(y_test, y_pred))
print("Autoencoder Confusion Matrix")
print(confusion_matrix(y_test, y_pred))
print("Autoencoder ROC-AUC Score:", roc_auc_score(y_test, mse))

# LSTM Autoencoder
print("\nLSTM Autoencoder Classification Report")
print(classification_report(y_test, y_pred_lstm))
print("LSTM Confusion Matrix")
print(confusion_matrix(y_test, y_pred_lstm))
print("LSTM ROC-AUC Score:", roc_auc_score(y_test, mse_lstm))

plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f'Autoencoder (AUC = {roc_auc:.2f})', color='darkorange')
plt.plot(fpr_lstm, tpr_lstm, label=f'LSTM (AUC = {roc_auc_lstm:.2f})', color='purple')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.show()

# Normalize GAN samples like original features
gan_scaled = scaler.transform(synthetic_attacks)

# Autoencoder
mse_gan_auto = np.mean(np.power(gan_scaled - autoencoder.predict(gan_scaled), 2), axis=1)
pred_auto_gan = [1 if e > threshold else 0 for e in mse_gan_auto]

# LSTM
gan_lstm = np.reshape(gan_scaled, (gan_scaled.shape[0], 1, gan_scaled.shape[1]))
mse_gan_lstm = np.mean(np.power(gan_lstm - model_lstm.predict(gan_lstm), 2), axis=(1,2))
pred_lstm_gan = [1 if e > threshold_lstm else 0 for e in mse_gan_lstm]

print("Autoencoder: % detected as attack:", np.mean(pred_auto_gan) * 100)
print("LSTM: % detected as attack:", np.mean(pred_lstm_gan) * 100)
```
